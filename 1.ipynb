1. Завдання полягає в тому, щоб побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 найпопулярніших пісень на Spotify y 2023 році за кількістю прослуховувань.


```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

data = pd.read_csv('spotify-2023.csv', encoding='cp1251')

data['streams'] = pd.to_numeric(data['streams'], errors='coerce')

data = data.dropna(subset=['streams'])

top_songs = data.nlargest(10, 'streams')

songs = top_songs['track_name'].tolist()
listens = top_songs['streams'].tolist()

plt.figure(figsize=(10, 6))
plt.barh(songs, listens, color='skyblue')
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Пісні')
plt.title('10 найпопулярніших пісень на Spotify у 2023 році')
plt.gca().invert_yaxis()
plt.show()
```

2. Розподіл жанрів найпопулярніших пісень: Створіть кругову діаграму або секторну діаграму, яка відображає розподіл жанрів серед найпопулярніших пісень у 2023 році.

```python


data = pd.read_csv('spotify-2023.csv', encoding='cp1251')

attributes = ['danceability_%', 'valence_%', 'energy_%']

data['combined_attribute'] = data[attributes].mean(axis=1)

plt.hist(data['combined_attribute'], bins=20, edgecolor='black')
plt.xlabel('Значення комбінованого атрибуту')
plt.ylabel('Частота')
plt.title('Розподіл комбінованих атрибутів')
plt.show()
```

3. Динаміка популярності треку в залежності від часу: Побудуйте лінійний графік, який відображає зміни кількості прослуховувань певного треку протягом року 2023 року.

```python


data = pd.read_csv('spotify-2023.csv', encoding='cp1251')

starboy_data = data[data['track_name'] == 'Starboy']

starboy_2023 = starboy_data[starboy_data['released_year'] == 2023]

plt.figure(figsize=(10, 6))
plt.plot(starboy_2023['released_month'], starboy_2023['streams'], marker='o', linestyle='-')
plt.title('Динаміка популярності треку "Starboy" протягом 2023 року')
plt.xlabel('Місяць')
plt.ylabel('Кількість прослуховувань')
plt.xticks(range(1, 13), ['Січ', 'Лют', 'Бер', 'Квіт', 'Трав', 'Черв', 'Лип', 'Серп', 'Вер', 'Жов', 'Лис', 'Груд'])
plt.grid(True)
plt.show()

```

4. Рейтинг артистів на Spotify: Створіть стовпчикову діаграму, на якій
відображаються 10 найпопулярніших артистів на Spotify у 2023 році за кількістю слухачів.

```python


data = pd.read_csv('spotify-2023.csv', encoding='cp1251')

data['streams'] = pd.to_numeric(data['streams'], errors='coerce')

top_artists = data.nlargest(10, 'streams')

plt.figure(figsize=(10, 6))
plt.bar(top_artists['artist(s)_name'], top_artists['streams'], color='skyblue')
plt.xlabel('Артист')
plt.ylabel('Кількість прослуховувань')
plt.title('10 найпопулярніших артистів на Spotify у 2023 році')
plt.xticks(rotation=45, ha='right')
plt.show()

```

5. Порівняння популярності трьох треків: Побудуйте горизонтальну
стовпчикову діаграму, яка порівнює кількість прослуховувань трьох конкретних треків у 2023 році.

```python


data = pd.read_csv('spotify-2023.csv', encoding='cp1251')

tracks = ['Seven (feat. Latto) (Explicit Ver.)', 'LALA', 'vampire']
tracks_data = data[data['track_name'].isin(tracks)]

plt.figure(figsize=(10, 6))
plt.barh(tracks_data['track_name'], tracks_data['streams'], color=['skyblue', 'lightgreen', 'salmon'])
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Назва треку')
plt.title('Порівняння популярності трьох треків на Spotify у 2023 році')
plt.show()

```
